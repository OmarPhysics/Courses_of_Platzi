### Librerías


```
import matplotlib.pyplot as plt   # librería para graficar
import numpy as np                # librería para manejo de vectores y utilidades matemáticas
```

# Funciones algebraicas

### Función lineal

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$




```
N = 100            # Puntos entre en dominio.
m = -5             # Pendiente
b = 5              # Ordenada al origen

# Definición de la función

def f(x):          
  return m*x+b

x = np.linspace(-15,15, num = N)     # Intervalo de dominio

y = f(x)           # Llamada a la función

# Gráfico

fig, ax = plt.subplots()            
ax.plot(x,y)
ax.grid()
```

### Funciones polinómicas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una función que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polinómicas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$





```
# Definición de la función

def f(x):           
  return (2*x**7) - x**4 + 3*x**2 + 4

y = f(x)                  # Llamada a la función

# Grafico

plt.plot(x,y)
plt.grid()
```

# Funciones trascendentes

Son funciones que no pueden ser expresadas con polinomios. 

### Funciones trigonométricas

Algunos ejemplos son las funciones $cos(x)$, $sen(x)$ y $tan(x)$

```
# Función seno

def f(x):
  return np.sin(x)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```
# Función coseno

def f(x):
  return np.cos(x)

y = f(x)

plt.plot(x,y)
plt.grid()
```

```
# Función tangente

def f(x):
  return np.tan(x)

y = f(x)

plt.plot(x,y)
plt.grid()
```

### Función exponencial

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una función exponencial es usando la base como el número de euler:

$$f(x)=e^x$$

```
# Función exponencial

def f(x):
  return np.exp(x)

y = f(x)

plt.plot(x,y)
plt.grid()
```

### Función logaritmo

El logaritmo está definido por la **relación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que está elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$





```
# Función logaritmo

def f(x):
  return np.log2(x)

x = np.linspace(0.001, 500, num = 2000)

plt.plot(x,f(x))
plt.grid()
```

# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```
# Función seccionada

def H(x):
  Y = np.zeros(len(x))
  for idx, x in enumerate(x):
    if x >= 0:
      Y[idx] = 1
  return Y

N=1000
x = np.linspace(-15, 15, num=N)
y = H(x)

# Gráfico

plt.plot(x,y)
plt.grid()
```

# Desplazamiento de funciones

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```
# Función original

N = 1000

def f(x):
  return x**2;

x = np.linspace(-10, 10, num = N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Función desplazada hacía arriba 20 unidades

N = 1000

def f(x):
  return x**2;

c = 20

x = np.linspace(-10, 10, num = N)

y = f(x) + c

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Función desplazada hacía abajo 20 unidades

N = 1000

def f(x):
  return x**2;

c = 20

x = np.linspace(-10, 10, num = N)

y = f(x) - c

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Función desplazada hacía la derecha 5 unidades

N = 1000

def f(x):
  return x**2;

c = 5

x = np.linspace(-15, 15, num = N)

y = f(x - c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Función desplazada hacía la izquierda 5 unidades

N = 1000

def f(x):
  return x**2;

c = 5

x = np.linspace(-15, 15, num = N)

y = f(x + c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=c \cdot f(x)$ alarga la gráfica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gráfica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gráfica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gráfica horizontelmente en un factor de $c$.

```
# Función original

N = 1000

def f(x):
  return np.sin(x);

x = np.linspace(-15, 15, num = N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Alargamiento vertical de la función

N = 1000

def f(x):
  return np.sin(x);

c = 5

x = np.linspace(-15, 15, num = N)

y = c*f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Compresión vertical de la función

N = 1000

def f(x):
  return np.sin(x);

c = 5

x = np.linspace(-15, 15, num = N)

y = (1/c)*f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Compresión horizontal de la función

N = 1000

def f(x):
  return np.sin(x);

c = 5

x = np.linspace(-15, 15, num = N)

y = f(c*x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

```
# Alargamiento horizontal de la función

N = 1000

def f(x):
  return np.sin(x);

c = 5

x = np.linspace(-15, 15, num = N)

y = f((1/c)*x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y = 0, color = 'r')
ax.axvline(x = 0, color = 'r')
```

## Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```
def H(x):
  Y = np.zeros(len(x))
  for idx, x in enumerate(x):
    if x >= 0:
      Y[idx] = 1
  return Y
  
y = H(x)

plt.plot(x,y)
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```
def f(x):
  return 1/(1 + np.exp(-x))

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```
def f(x):
  return np.tanh(x)
  
y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función ReLU

$$R(x)=max(0,x)$$

```
def f(x):
  return np.maximum(x,0)

y = f(x)

plt.plot(x,y)
plt.grid()
```
